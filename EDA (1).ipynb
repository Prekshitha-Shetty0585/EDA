```python
#Data Cleaning and Preparation:  
import pandas as pd
import numpy as np
import matplotlib as plt
from scipy import stats
data = pd.read_csv(r"C:\Users\Prekshitha Shetty\EDA.csv")
data
```

```python
data.info()
```

```python
data.dtypes
```

```python
data.nunique()
```

```python
##Handle missing values appropriately
```

```python
#check miss values
data.isnull().sum()
# Handle missing values (e.g., imputation with mean or median, or deletion)
data = data.fillna(data.median())
```

```python
##### Identify and correct inconsistencies in data types
```

```python
# Check data types
print(data.dtypes)
# Convert any numerical values stored as strings to numerical types
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = pd.to_numeric(data[column], errors='coerce')
```

```python
#### Detect and treat outliers
```

```python
# Detect outliers using Z-score
from scipy.stats import zscore
z_scores = np.abs(zscore(data))
outliers = (z_scores > 3).all(axis=1)
data = data[~outliers]
```

```python
## 2. Statistical Summary
#### Provide a statistical summary
```

```python
# Statistical summary for each variable
summary = data.describe().transpose()
summary['median'] = data.median()
summary['IQR'] = summary['75%'] - summary['25%']
# Display the summary
print(summary)
```

```python
## 3. Data Visualization
#### Create histograms or boxplots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Histograms
data.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
# Boxplots
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.show()
```
